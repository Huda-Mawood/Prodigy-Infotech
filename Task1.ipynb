Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import Data CSV File

```python
df=pd.read_csv("population_by_country_2020.csv")
```

```python
df.head()
```

Prepare Data

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df=df.dropna()
df.shape
```

```python
df=df.drop_duplicates()
df.shape
```

```python
df["Med. Age"].max()
```

```python
df["Yearly Change"]=df["Yearly Change"].str.replace('%',"").astype(float)
df["Yearly Change"]
```

```python
dfObserve=df[["Country (or dependency)","Population (2020)","Yearly Change","Med. Age"]]
```

```python
dfObserve
```

```python
maxYearlyChange=dfObserve["Yearly Change"].max()
maxYearlyChange
```

```python
minYearlyChange=dfObserve["Yearly Change"].min()
minYearlyChange
```

```python
maxPopulation=dfObserve["Population (2020)"].max()
maxPopulation
```

```python
minPopulation=dfObserve["Population (2020)"].min()
minPopulation
```

```python
dfObserve.sort_values(["Population (2020)","Yearly Change"],ascending=[False,True])
```

```python
dfObserve[dfObserve["Yearly Change"]==maxYearlyChange]
```

```python
dfObserve[dfObserve["Yearly Change"]==minYearlyChange]
```

```python
dfObserve[dfObserve["Population (2020)"]==maxPopulation]
```

```python
dfObserve[dfObserve["Population (2020)"]==minPopulation]
```

visualize Data

```python
fig, ax = plt.subplots(figsize=(15, 6))
plt.bar(dfObserve["Yearly Change"],dfObserve["Med. Age"],width=.05);
plt.xlabel("Yearly Change")
plt.ylabel("Med. Age")
plt.title("Yearly change in Median Age");
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
plt.hist(dfObserve["Med. Age"]);
plt.xlabel("Med. Age")
plt.ylabel("frequency")
plt.title(" Distribution the Median Age")
```
